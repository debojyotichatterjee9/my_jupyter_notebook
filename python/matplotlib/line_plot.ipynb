```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
x_axis = np.arange(0,10,1)
print(x_axis)
```

```python
y_axis = random.sample(range(1, 50), 10)
print(y_axis)
```

```python
plt.plot(x_axis,y_axis)
plt.xlabel("Days")
plt.ylabel("Occurence")
plt.title("Random Graph Data")
```
